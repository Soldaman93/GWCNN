```python
from tqdm import tqdm

import datetime

import pandas as pd

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Reshape, GlobalAveragePooling1D
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv1D, MaxPooling1D, ReLU, Softmax
from tensorflow.keras.models import model_from_json

import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
from random import seed
from random import random
```

```python
# Setting the parameters

seed(2) # setting the seed for random()
np.random.seed(2) # setting the seed for np.random.normal()
tf.random.set_seed(2) # setting the seed for tensorflow

n_timesteps = 300 # number of time (x) points

n_sin_train_samples = 1000 # number of sin samples used for training/validation
n_noise_train_samples = 1000 # number of noise samples used for training/validation

x = np.linspace(0,10*np.pi,n_timesteps) # timesteps

omega=np.zeros((1))
phase=np.zeros((1))

noise_mean = 0.0 # mean of the gaussian noise
noise_std = 1.0 # standard deviaiton of the gaussian noise

################ Training/validation set ################

sin_signal=np.zeros((n_sin_train_samples,n_timesteps+1))
noise_signal=np.zeros((n_noise_train_samples,n_timesteps+1))


for i in range(0,n_sin_train_samples):
    omega = 0.5 + 3*random() # frequency of a single signal # tried: 0.5, 1.0, 2.0, 4.0
    phase = 2*np.pi*random() # phase of a single signal
    sin_amplitude = 0.5 + 0*random()*x/np.max(x) # amplitude of a single signal
    sin_noise = np.random.normal(noise_mean, noise_std, size=(n_timesteps)) # noise to be added to the signal
    for j in range(0,n_timesteps):
        sin_signal[i,j] = (sin_amplitude[j]*np.sin(omega*x[j]+phase) + sin_noise[j]) # signal + noise samples
    sin_signal[i,:]/=np.max(np.abs(sin_signal[i,:]))
    
for i in range(0,n_noise_train_samples):
    for j in range(0,n_timesteps):
        noise_signal[i,j]=(np.random.normal(noise_mean, noise_std, size=(n_timesteps))[j]) # pure noise samples
    noise_signal[i,:]/=np.max(np.abs(noise_signal[i,:]))

for i in range(0,n_sin_train_samples):
    sin_signal[i,-1]=np.int(1) # 1 means there is a signal
    
for i in range(0,n_noise_train_samples):
    noise_signal[i,-1]=np.int(0) # 0 means there is only noise
    
signals=np.vstack((sin_signal,noise_signal))
np.random.shuffle(signals) # shuffling all the samples

x_train = np.delete(signals, (-1), axis=1)
y_train = signals[:,-1]

x_train = np.asarray(x_train, dtype= np.float32).reshape(-1, n_timesteps, 1)
y_train = np.asarray(y_train)

print("\n --- Training/validation set shape --- \n")
print("x:", x_train.shape)
print("y:", y_train.shape)
```

```python
fig = go.Figure()

for i in range(5,10):
    fig.add_trace(go.Scatter(x=x, y=signals[i,:],
                        mode='lines',
                        name=signals[i,-1]))

fig.show()
#print(signals[1,:])
```

```python
input_shape = (n_timesteps)
num_classes = 1

tf.keras.backend.clear_session() # clearing the previous session

print("\n --- Creating convolutional neural network model --- \n")

model_m = Sequential()

activ_func='relu'

model_m.add(Conv1D(100, 10, activation=activ_func, input_shape=(n_timesteps, 1)))
model_m.add(Conv1D(100, 10, activation=activ_func))
model_m.add(MaxPooling1D(3))
model_m.add(Conv1D(100, 10, activation=activ_func))
model_m.add(Conv1D(100, 10, activation=activ_func))
model_m.add(GlobalAveragePooling1D())
model_m.add(Dropout(0.5))
model_m.add(Dense(num_classes, activation='sigmoid'))

print(model_m.summary())
```

```python
%tensorboard --logdir='./logs' --port=8008
```

```python
print("\n --- Training convolutional neural network model --- \n")

# %rm -r logs # removing the old logs
# %rm best_model*

callbacks_list = [
   keras.callbacks.ModelCheckpoint(
       filepath='best_model.{epoch:02d}-{val_binary_accuracy:.2f}.h5',
       monitor='val_binary_accuracy', save_best_only=True),
    keras.callbacks.EarlyStopping(monitor='val_binary_accuracy', patience=5000),
    keras.callbacks.TensorBoard(
    log_dir='./logs', histogram_freq=1, write_graph=True, write_images=True,
    update_freq='epoch')
]

model_m.compile(loss='binary_crossentropy',
                optimizer='Adam', metrics=['binary_accuracy'])

BATCH_SIZE = 1
EPOCHS = 20

history = model_m.fit(x=x_train,
                      y=y_train,
                      batch_size=BATCH_SIZE,
                      epochs=EPOCHS,
                      callbacks=callbacks_list,
                      validation_split=0.4,
                      verbose=1)
```

```python
print("\n --- Learning curve of model training --- \n")

# summarize history for accuracy and loss

fig = go.Figure()


fig.add_trace(go.Scatter(x=np.linspace(0,EPOCHS), y=history.history['binary_accuracy'],
                        line = dict(color='green', width=2, dash='dot'),
                        name='Accuracy of training data'))

fig.add_trace(go.Scatter(x=np.linspace(0,EPOCHS), y=history.history['val_binary_accuracy'],
                        line = dict(color='green', width=2),
                        name='Accuracy of validation data'))

fig.add_trace(go.Scatter(x=np.linspace(0,EPOCHS), y=history.history['loss'],
                        line = dict(color='red', width=2, dash='dot'),
                        name='Loss of training data'))

fig.add_trace(go.Scatter(x=np.linspace(0,EPOCHS), y=history.history['val_loss'],
                        line = dict(color='red', width=2),
                        name='Loss of validation data'))


fig.update_layout(title='Model accuracy and loss',
                   xaxis_title='Epoch',
                   yaxis_title='Accuracy and loss',
                   width=1000,
                   height=500,
                   yaxis_type="log")

fig.show()
```

```python
# Save model to JSON file

model_json = model_m.to_json()
with open("my_cnn.json", "w") as json_file:
    json_file.write(model_json)

# Save weights to HDF5 file

model_m.save_weights("my_cnn.h5")

print("Saved trained model to disk")
```

```python
# Load JSON and create model

json_file = open('my_cnn.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

# Load weights into new model

loaded_model.load_weights("my_cnn.h5")
print("Loaded trained model from disk")
```

```python
################ Test set ################

seed(3) 
np.random.seed(3) 
tf.random.set_seed(3) 

n_sin_test_samples = 1000 # number of sin samples used for testing
n_noise_test_samples = 1000 # number of noise samples used for testing

sin_signal=np.zeros((n_sin_test_samples,n_timesteps+1))
noise_signal=np.zeros((n_noise_test_samples,n_timesteps+1))

noise_mean = 0.0 
noise_std = 1.0 

for i in range(0,n_sin_test_samples):
    omega = 0.5 + 3*random() 
    phase = 2*np.pi*random() 
    sin_amplitude = 0.2 + random()
    sin_noise=np.random.normal(noise_mean, noise_std, size=(n_timesteps))
    for j in range(0,n_timesteps):
        sin_signal[i,j] = sin_amplitude*np.sin(omega*x[j]+phase) + sin_noise[j]
    sin_signal[i,:]/=np.max(np.abs(sin_signal[i,:]))

for i in range(0,n_noise_test_samples):
    for j in range(0,n_timesteps):
        noise_signal[i,j]=np.random.normal(noise_mean, noise_std, size=(n_timesteps))[j]
    noise_signal[i,:]/=np.max(np.abs(noise_signal[i,:]))
        
for i in range(0,n_sin_test_samples):
    sin_signal[i,-1]=1 # 1 means signal
    
for i in range(0,n_noise_test_samples):
    noise_signal[i,-1]=0 # 0 means noise
    
signals=np.vstack((sin_signal,noise_signal))
np.random.shuffle(signals)

x_test = np.delete(signals, (-1), axis=1)
y_test = signals[:,-1]

x_test = np.asarray(x_test, dtype= np.float32).reshape(-1, n_timesteps, 1)
y_test = np.asarray(y_test)

print("\n --- Test set shape --- \n")
print("x:", x_test.shape)
print("y:", y_test.shape)
```

```python
# Check against test data

print("\n --- Checking against test data --- \n")

# loaded_model=model_m

loaded_model.compile(loss='binary_crossentropy',
                optimizer='adam', metrics=['binary_accuracy'])
score = loaded_model.evaluate(x_test, y_test, verbose=1)

print("\nAccuracy on test data: %0.4f" % score[1])
print("\nLoss on test data: %0.4f" % score[0])
```

```python
# To test a single signal

# seed(4) 
# np.random.seed(4) 
# tf.random.set_seed(4) 

n_timesteps = 300

x = np.linspace(0,10*np.pi,n_timesteps) 

noise_mean = 0.0 
noise_std = 1.0 

omega = 0.5 + 3*random() 
phase = 2*np.pi*random() 
sin_amplitude = 0.2 + random()
 
sin_noise = np.random.normal(noise_mean, noise_std, size=(n_timesteps)) # noise to be added to the signal
sin_signal = sin_amplitude*np.sin(omega*x+phase) + sin_noise # signal + noise sample

x_test_single = np.asarray(sin_signal, dtype= np.float32).reshape(-1, n_timesteps, 1)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=sin_signal-sin_noise,
                    mode='lines',
                    name='Pure signal'))

fig.add_trace(go.Scatter(x=x, y=sin_signal,
                    mode='lines',
                    name='Noisy signal'))

fig.update_layout(title='Input signal',
                   xaxis_title='Timestep',
                   yaxis_title='Signal')

fig.show()

signal_power = np.sum((sin_signal-sin_noise-np.mean(sin_signal-sin_noise))**2)**0.5
noise_rms = np.sum((sin_noise-np.mean(sin_noise))**2)**0.5/(n_timesteps)**0.5

SNR = np.max(signal_power)/noise_rms

print("\n SNR of the input signal:",SNR)

print("\n Predicted label of the input signal (1=signal; 0=noise):",(loaded_model.predict(x_test_single) > 0.5).astype("int32")[0][0])
```
